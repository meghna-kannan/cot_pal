```python
import re
def extract_cot(cot_answer: str) :
    try :
        if not (cot_answer is None or cot_answer == "" or cot_answer == "cot error") :
            last_line = cot_answer.strip().split('\n')[-1]
            #if 'So the answer is' in last_line:
            pattern = r"\d+(?:,\d+)*(?:\.\d+)?"
            numbers = re.findall(pattern, last_line)
            numbers = [int(num.replace(',', '')) if '.' not in num else float(num.replace(',', '')) for num in numbers]
            #numbers = [num.replace(',', '') for num in numbers]
            return numbers[0]
        else : 
            return -100
    except Exception as e:
        print("An exception occurred:", e)
        print(cot_answer)
        print()
        return -100
```

```python
import signal

def timeout_handler(signum, frame):
    raise TimeoutError("Execution timed out")

def execute_with_timeout(code, timeout):
    signal.signal(signal.SIGALRM, timeout_handler)
    signal.alarm(timeout)

    try:
        exec(code, locals())
        result = locals()['solution']()
        return result
    finally:
        signal.alarm(0)  # Cancel the alarm
```

```python
def extract_pal(pal_answer: str) :
    try :
        if not (pal_answer is None or pal_answer == "" or pal_answer == "cot error") :
            result = execute_with_timeout(pal_answer, 5)
            return result
        else : 
            return pal_answer
    except TimeoutError as e:
        print("Execution timed out", e)
        #print(pal_answer)
        return -100
    except Exception as e:
        print("An exception occurred:", e)
        #print(pal_answer)
        #print()
        return -100
```

```python
import os

folder_path = 'result/gsm8K'
file_pattern = '.jsonl'  # Example pattern, change it to match your desired pattern

cot_correct = 0
cot_wrong = 0
cot_error = 0

pal_correct = 0
pal_wrong = 0
pal_error =0 
both_correct = 0

for file_name in os.listdir(folder_path):
    if file_name.endswith(file_pattern):
        lines = jsonlines_load_file(f'{folder_path}/{file_name}')
        for line in lines :
            print('Processing ' + str(line['index']))
            answer = line['answer']
            cot = line['cot']
            pal = line['pal']
            cot_answer = None
            pal_answer = None
           
            if cot is None or cot == "" or cot == "cot error":
                cot_error+=1
            else : 
                cot_answer = extract_cot(cot)
                if (answer == cot_answer):
                    cot_correct+=1
                else :
                    #print('cot wrong' + str(answer) + ' ' + str(cot_answer))
                    cot_wrong+=1
                
            if pal is None or pal == "" or pal == "pal error":
                pal_error+=1
            else : 
                pal_answer = extract_pal(pal)
                if (answer == pal_answer):
                   pal_correct+=1
                else :
                    #print('pal wrong' + str(answer) + ' ' + str(pal_answer))
                    pal_wrong+=1
            if not (cot_answer is None or pal_answer is None):
                if (answer == cot_answer and answer == pal_answer):
                    both_correct+=1

            dict = { 
                'index' : line['index'],
                'question' : line['question'],
                'answer' : line['answer'],
                'cot': cot_answer,
                'pal': pal_answer,
                'cot_sol' : cot,
                'pat_sol' : pal
            }
            append_json_to_file('analysis', 'gsm8K-0301', dict)
            
```
