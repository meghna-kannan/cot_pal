````python
import os
import ipywidgets as widgets

folder_path = 'analysis'
file_pattern = 'svamp-all.jsonl'  # Example pattern, change it to match your desired pattern


# Create the dropdown widget

cot_correct=0
cot_wrong=0
cot_correct_pal_wrong=0
pal_correct=0
pal_wrong=0
pal_correct_cot_wrong=0
both_correct=0
both_wrong=0

for file_name in os.listdir(folder_path):
    if file_name.endswith(file_pattern):
        lines = jsonlines_load_file(f'{folder_path}/{file_name}')
        for line in lines :
            if (line['answer']== line['cot'] and line['answer'] == line['pal']):
                both_correct = both_correct + 1

            if (line['answer']!= line['cot'] and line['answer'] != line['pal']):
                both_wrong = both_wrong + 1

            if (line['answer']== line['cot']):
                cot_correct = cot_correct + 1

            if (line['answer']!= line['cot']):
                 cot_wrong = cot_wrong + 1

            if (line['answer']== line['cot'] and line['answer'] != line['pal']):
                 cot_correct_pal_wrong = cot_correct_pal_wrong + 1

            if (line['answer']== line['pal']):
                pal_correct = pal_correct + 1

            if (line['answer']!= line['pal']):
                 pal_wrong = pal_wrong + 1

            if (line['answer']== line['pal'] and line['answer'] != line['cot']):
                 pal_correct_cot_wrong = pal_correct_cot_wrong + 1


print(f' both_correct {both_correct}')
print(f' both_wrong {both_wrong}')
print(f' cot_correct {cot_correct}')
print(f' cot_wrong {cot_wrong}')
print(f' cot_correct_pal_wrong {cot_correct_pal_wrong}')
print(f' pal_correct {pal_correct}')
print(f' pal_wrong {pal_wrong}')
print(f' pal_correct_cot_wrong {pal_correct_cot_wrong}')







            

            
                
                
            
              
````
